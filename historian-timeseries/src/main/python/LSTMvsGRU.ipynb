```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
```

```python
gru_id = read_csv('GRU/GRU_id.csv')
gru_name = read_csv('GRU/GRU_name.csv')
lstm_name = read_csv('LSTM/LSTM_name.csv')
lstm_id = read_csv('LSTM/LSTM_id.csv')
```

```python
moy_gru_id, moy_lstm_id, moy_gru_name, moy_lstm_name = 0, 0, 0, 0
for i in range(0, gru_id.shape[0]):
    moy_gru_id += gru_id.loc[i,'mean_squarred_error']
    moy_lstm_id += lstm_id.loc[i,'mean_squarred_error']
    moy_gru_name += gru_name.loc[i,'mean_squarred_error']
    moy_lstm_name += lstm_name.loc[i,'mean_squarred_error']
print('gru_id : ', moy_gru_id/gru_id.shape[0], '\nlstm_id : ', moy_lstm_id/gru_id.shape[0],'\ngru_name : ', moy_gru_name/gru_id.shape[0], '\nlstm_name : ', moy_lstm_name/gru_id.shape[0])
```

```python
fig=plt.figure(figsize=(12,4), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(gru_name['mean_squarred_error'].to_numpy().reshape(-1,1))
plt.plot(gru_id['mean_squarred_error'].to_numpy().reshape(-1,1))
plt.plot(lstm_name['mean_squarred_error'].to_numpy().reshape(-1,1))
plt.plot(lstm_id['mean_squarred_error'].to_numpy().reshape(-1,1))
fig.legend(['GRU_name', 'GRU_id', 'LSTM_name', 'LSTM_id'])
plt.title('Root mean squarred error of the different methods')
plt.show()
```
